```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
```

```python
df = pd.read_excel('Dataset_Data_Mining_Completo_5000.xlsx')
```

```python
dados = df[['Idade', 'Renda_Mensal', 'Valor_Total']]
```

```python
modelo = KMeans(n_clusters=3, random_state=42)
modelo.fit(dados)
```

```python
df['Grupo'] = modelo.labels_
```

```python
df.head(10)
```

```python
df.groupby('Grupo').mean(numeric_only=True)
```

```python
x = df[['Idade', 'Renda_Mensal', 'Valor_Total', 'Avaliacao']]

y = df['Cancelou']
```

```python
arvore = DecisionTreeClassifier(random_state=42, max_depth=4)
```

```python
arvore.fit(x, y)
```

```python
plt.figure(figsize=(20, 10))

plot_tree(
    arvore,
    feature_names=x.columns,
    class_names=arvore.classes_,
    rounded=True,
    fontsize=9
    )
```
